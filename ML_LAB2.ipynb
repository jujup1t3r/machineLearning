**1.5 Display Missing Value**  
Display the count and percentage of missing values for each column.

```python
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_summary = pd.DataFrame({'missing_count': missing, 'missing_pct(%)': missing_pct})
missing_summary = missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_count', ascending=False)
missing_summary
```

Generate a bar chart to visualize the count of missing values.

```python
fig, ax = plt.subplots(figsize=(3, 3))
sns.barplot(x=missing_summary.index, y=missing_summary['missing_count'], ax=ax, color='#e76f51')
ax.set_title('Missing Values', fontsize=13, fontweight='bold')
ax.set_ylabel('Row')
ax.set_xlabel('')
for i, v in enumerate(missing_summary['missing_count']):
    ax.text(i, v + 5, f'{v:,}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

**1.6 Display Duplicate Records**  
Check for duplicate records across all columns to prevent data redundancy that could affect the analysis.

**1.7 Display Class Distribution**  
Analyze the distribution of the target variable to check for class imbalance within the dataset.

```python
class_counts = df['Heart_Disease'].value_counts()
class_percentages = df['Heart_Disease'].value_counts(normalize=True) * 100

print("Class Counts:")
print(class_counts)
print("\nClass Percentages (%):")
print(class_percentages)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))
sns.countplot(x='Heart_Disease', data=df, ax=ax, color='#e76f51')
ax.set_title('Class Distribution', fontsize=13, fontweight='bold')
ax.set_xlabel('Heart Disease (0 = No, 1 = Yes)')
ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

**LAB2: Data Visualization**

**2.1 Histogram**  
This section presents side-by-side visualizations to analyze missing value distribution, numerical features (BMI), and categorical risk factors (Smoking Status) within the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

sns.barplot(x=missing_summary.index, y=missing_summary['missing_count'], ax=axes[0], color='#e76f51')
axes[0].set_title('Missing Values', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].set_xlabel('')

sns.histplot(data=df, x='BMI', kde=True, ax=axes[1], color='#2a9d8f', bins=20)
axes[1].set_title('BMI Distribution', fontsize=11, fontweight='bold')
axes[1].set_xlabel('BMI')
axes[1].set_ylabel('Count')

sns.countplot(x='Smoking', data=df, ax=axes[2], color='#e9c46a', 
              order=df['Smoking'].value_counts().index)
axes[2].set_title('Smoking Status', fontsize=11, fontweight='bold')
axes[2].set_xlabel('Smoking Habit')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

**2.2 Corelation Heatmap**  
Generate a correlation heatmap to analyze the linear relationships between all numerical features and identify potential predictors for the target variable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = df.corr(numeric_only=True)

fig, ax = plt.subplots(figsize=(8, 7))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', 
            square=True, linewidths=0.5, ax=ax, annot_kws={"size": 8})

ax.set_title('Correlation Heatmap of Numerical Features', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```
